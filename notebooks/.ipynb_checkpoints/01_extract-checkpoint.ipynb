# 01 - Extração de Dados da Open Brewery API (Bronze Layer)

Acessar a API pública do [Open Brewery DB](https://www.openbrewerydb.org/)
Coletar os dados de cervejarias.
Armazenar os dados brutos em formato `.json` na camada **bronze**

```python
import os
import requests
import pandas as pd
from datetime import datetime
```

```python
url = "https://api.openbrewerydb.org/v1/breweries"

response = requests.get(url)
print("STATUS:", response.status_code)

if response.status_code == 200:
    data = response.json()
    print()
    print("JSON EXPORTADO COM SUCESSO !!!")
    print()
    for brewery in data[:3]:
        print(brewery)
else:
    print("ERRO") 
```

```python
df = pd.DataFrame(data)
df.head()
```

```python
file_name = f"data/bronze/breweries_{datetime.now().date()}.json"
df.to_json(file_name, orient='records',indent=2)
print(f"Dados Salvos com Sucesso em: {file_name}")
```
